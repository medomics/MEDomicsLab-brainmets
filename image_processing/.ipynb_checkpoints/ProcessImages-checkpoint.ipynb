```python
import sys
import os
import pydicom
from pathlib import Path
import scipy.io


import matplotlib.pyplot as plt
import time
import pandas as pd
import csv
import math
import numpy as np
import glob

import helper
```

```python
IMAGES = '/data/public/MIM_BMETS_V6/1_images_masks_raw/images/'

npImages = {}
c=0
fnamesImages = glob.glob(IMAGES + '/' + '*.npy')
fnamesImages.sort()
for np_name in fnamesImages:
    npImages[c] = np.load(np_name)
    c = c + 1;
```

```python
brainMASKS = '/data/public/MIM_BMETS_V6/1_images_masks_raw/brain_masks/'

npBrainMasks = {}
c=0
fnamesBrainMasks = glob.glob(brainMASKS + '/' + '*.npy')
fnamesBrainMasks.sort()
for np_name in fnamesBrainMasks:
    npBrainMasks[c] = np.load(np_name)
    c = c + 1;
```

```python
metsMASKS = '/data/public/MIM_BMETS_V6/1_images_masks_raw/mets_masks'

npMetsMasks = {}
c=0
fnamesMetsMasks = glob.glob(metsMASKS + '/' + '*.npy')
fnamesMetsMasks.sort()
for np_name in fnamesMetsMasks:
    npMetsMasks[c] = np.load(np_name)
    c = c + 1;
```

```python
# need to check i==27, BrainMets-UCSF-00268_20091227, NotImplementedError: multi-dimensional slicing is not implemented
# need to check i==60, BrainMets-UCSF-00572_20130221
# image size might by 512x512 rather than 256x256, and one that has 704

for i in range(0,10):
#     if i==27 or i==60:
#         continue
        
    print(fnamesBrainMasks[i])
    print(fnamesImages[i])
    skullStripped0 = helper.extract_brain(npImages[i], npBrainMasks[i])
    skullStripped1 = helper.z_pad(skullStripped0)
    skullStripped2 = helper.xy_pad(skullStripped1)
    skullStripped3 = helper.normalize(skullStripped2, npBrainMasks[i])
    
    t = fnamesImages[i].split('/')
    np.save('/data/public/MIM_BMETS_V6/2_images_masks_processed/skull_stripped/' + t[-1], skullStripped3)
    continue
    
    image = helper.z_pad(npImages[i])
    image = helper.xy_pad(image)
    image = helper.normalize(image, npBrainMasks[i])
    
    t = fnamesImages[i].split('/')
    np.save('/data/public/MIM_BMETS_V6/2_images_masks_processed/images/' + t[-1], image)
    
    metsMask = helper.z_pad(npMetsMasks[i])
    metsMask = helper.xy_pad(metsMask)
    
    t = fnamesMetsMasks[i].split('/')
    np.save('/data/public/MIM_BMETS_V6/2_images_masks_processed/mets_masks/' + t[-1], metsMask)
    
    brainMask = helper.z_pad(npBrainMasks[i])
    brainMask = helper.xy_pad(brainMask)
    
    t = fnamesBrainMasks[i].split('/')
    np.save('/data/public/MIM_BMETS_V6/2_images_masks_processed/brain_masks/' + t[-1], brainMask)
            
```
