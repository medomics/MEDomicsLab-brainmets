```python
totStudies = 0
allMetsROIs = []
totMets = 0
all_dirs = ['2015-06_2__Studies']#dirs # skipped 2011-03__Studies
#all_dirs = ['1997-07__Studies']

notMetsROIs = ['{Brain}','{Ventricles}','{*Skull}','{optic}','{motor}','{cord}','{External}','{ISOCENTER}', \
             '{avoid}','{Fornix}','{optic_n}','{brain_4gy}','{brain_4GY}','{l-motor}','{brainstem}', \
             '{intcaps}','{ON}','{cyst3}','{olfact}','{chiasm}','{cavity}','{L_optic_n}', \
             '{Avoidance}','{opticpath}','{right_hc}','{old_set}','{test}','{ROI}', \
             '{avoid motor}','{Avoid}','{L optic n}','{14&16}','{2,3,4}','{Target 7}','{Target 6}']

for each_dir in all_dirs:
    print(each_dir)
    
    # need to figure out why these studies are not like the others
    # 19970708 has '*Skull' and 'targets' as roi for met
    #if each_dir=='1997-07__Studies':
    #    continue
    
    temp = os.listdir(images_dir + '/' + each_dir)
    if '.DS_Store' in temp:
        temp.remove('.DS_Store')
    
    temp.sort()

    #print( len(temp) , temp[0], temp[1], temp[2], temp[3])
    #if len(temp) % 4 != 0:
    #print(len(temp))

    # each imaging study should have 4 files
    # 1st file should be MR, 4th should be RTstruct
    d = 0
    for i in range(0,2*len(temp),4):
        # skip incomplete studies
        if (len(temp) - (i-d) ) < 4:
            break        
        
        # a bunch of BS to get MR ans RTstrcut files
        i = i - d
        n0 = temp[i].split('_')
        n1 = temp[i+1].split('_')
        n2 = temp[i+2].split('_')
        n3 = temp[i+3].split('_')
        #print(i)
        #print(n0)
        #print(n1)
        #print(n2)
        #print(n3)
        
        # two different study name types: 1) starts with 'Brain', 2) starts with numbers
        if n0[0][0]=='B':
            # sort based on date
            print('here')
            a = [i.split('_') for i in temp]
            b = [row[3] + '_' + row[4] for row in a]
            temp = [x for _,x in sorted(zip(b,temp))]
            print(temp)
            n0 = temp[i].split('_')
            n1 = temp[i+1].split('_')
            n2 = temp[i+2].split('_')
            n3 = temp[i+3].split('_')
            if n0[0] != n1[0] :
                d = d + 3
                #print('A0')
                continue
            elif n0[0] != n2[0]:
                d = d + 2
                #print('A1')
                continue
            elif n0[0] != n3[0]:
                d = d + 1
                #print('A2')
                continue


        else:
            if n0[1] != n1[1] :
                d = d + 3
                #print('A0')
                continue
            elif n0[1] != n2[1]:
                d = d + 2
                #print('A1')
                continue
            elif n0[1] != n3[1]:
                d = d + 1
                #print('A2')
                continue

        print('MR file name',temp[i])
        #if temp[i] != '2.16.840_552.8054_MR_2003-03-11_000000_._T1CE_n49__00000':
        #    continue
            
        # read RTstruct, get number of ROIs
        temp2 = os.listdir(images_dir + '/' + each_dir + '/' + temp[i+3])
        if '.DS_Store' in temp2:
            temp2.remove('.DS_Store')
        temp2.sort()
        rtFileName = temp2[0]
        f = pydicom.read_file(images_dir + '/' + each_dir + '/' + temp[i+3] + '/' + rtFileName, force=True)
        
        # StructureSetROISequence, number of ROIs
        structROIs = f.StructureSetROISequence
        nROIs = int(structROIs[-1].ROINumber)
        
        print('Number of rois: ', structROIs[-1].ROINumber)
        
        # copy struct file into same dir as images for MEDimage to read
        src = images_dir + '/' + each_dir + '/' + temp[i+3] + '/' + rtFileName
        dst = images_dir + '/' + each_dir + '/' + temp[i] + '/' + rtFileName
        copyfile(src, dst)
        
        MASKSdata = None
        s = 0
        # go through each ROI and sum masks of mets
        for j in range(nROIs):
            # get roi name and id
            rtPath = Path(images_dir + '/' + each_dir + '/' + temp[i])
            print(rtPath)
            roi = '{' + structROIs[j].ROIName + '}'
            roiN = structROIs[j].ROINumber

            # skip roi if it is not a targeted met
            if roi != '{Brain}':#roi in notMetsROIs or roi[1]=='*':
                continue
            # sum mets rois into 1 mask (MASKSdata)
            else:
                allMetsROIs.append(roi) # save list of ROIs for review
                totMets = totMets + 1
                
                print(j, "reading the following roi:", roi )
                MEDIobj = MEDimage(path_patient=rtPath, name_roi=str(roiN -1), compute_radiomics_features=False, save_modified_image=True)

                MASK = None
                IMAGE, MASK = MEDIobj.get_image_and_mask()

                if s == 0:
                    MASKSdata = MASK.data
                    s = 1
                else:
                    MASKSdata = MASKSdata.__add__(MASK.data)
    
        
        # save MASKS for pateint

        study_name = f.PatientID + '_' + f.StudyDate 
        #file_name = '/data/public/MIM_BMETS_V6/1_images_masks_raw/images/' + study_name + '_image'
        #np.save(file_name, IMAGE.data)
        #file_name = '/data/public/MIM_BMETS_V6/1_images_masks_raw/mets_masks/' + study_name + '_mask'
        #np.save(file_name, MASKSdata)
        file_name = '/data/public/MIM_BMETS_V6/1_images_masks_raw/brain_masks/' + study_name + '_brain_mask'
        np.save(file_name, MASKSdata)

        totStudies = totStudies + 1
        print('STUDY NAME:',study_name)
    
```
